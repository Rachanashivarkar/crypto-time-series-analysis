```python
import yfinance as yf
import pandas as pd
```

```python
data = yf.download(
    "BTC-USD",
    start="2024-01-01",
    end="2026-01-01"
)
```

```python
df = data.reset_index()

df = df[["Date","Open","High","Low","Close","Volume"]]
```

```python
df.columns = ["timestamp","open","high","low","close","volume"]

import os
os.makedirs("data/raw", exist_ok=True)
df.to_csv("data/raw/bitcoin_ohlcv.csv", index=False)
print("Bitcoin OHLCV data collected successfully")
```
