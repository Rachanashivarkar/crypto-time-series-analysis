```python
import pandas as pd

arima_pred = pd.read_csv("./../results/arima_predictions.csv")
sarima_pred = pd.read_csv("./../results/sarima_predictions.csv")
prophet_pred = pd.read_csv("./../results/prophet_predictions.csv")
lstm_pred = pd.read_csv("./../results/lstm_predictions.csv")

data = pd.read_csv("data/processed/bitcoin_processed.csv")
```

```python
test = data["close"].tail(68)
```

```python
arima_values = arima_pred["prediction"]

sarima_values = sarima_pred["prediction"]

prophet_values = prophet_pred["yhat"].tail(len(test))

lstm_values = lstm_pred["prediction"]

test_lstm = test[-len(lstm_values):]
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12,6))

plt.plot(test.values, label="Actual Price", linewidth=3)

plt.plot(arima_values.values, label="ARIMA", linestyle="--")

plt.plot(sarima_values.values, label="SARIMA")

plt.plot(prophet_values.values, label="Prophet")

plt.plot(range(len(test)-len(lstm_values), len(test)), lstm_values.values, label="LSTM", linewidth=3)

plt.title("Bitcoin Price Forecast Comparison")

plt.xlabel("Time")
plt.ylabel("Price")

plt.legend()
plt.grid(True)

plt.show()
```
