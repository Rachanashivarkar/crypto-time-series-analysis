```python
import numpy as np
import pandas as pd

from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_percentage_error
```

```python
df = pd.read_csv("data/processed/bitcoin_processed.csv")

df["timestamp"] = pd.to_datetime(df["timestamp"])

df.set_index("timestamp", inplace=True)

data = df["close"]
```

```python
train = data[:int(len(data)*0.8)]
test = data[int(len(data)*0.8):]
```

```python
arima_pred = pd.read_csv("./../results/arima_predictions.csv")
```

```python
sarima_pred = pd.read_csv("../results/sarima_predictions.csv")
```

```python
prophet_pred = pd.read_csv("../results/prophet_predictions.csv")
```

```python
lstm_pred = pd.read_csv("../results/lstm_predictions.csv")
```

```python
mae_arima = mean_absolute_error(test, arima_pred["prediction"])

rmse_arima = np.sqrt(mean_squared_error(test, arima_pred["prediction"]))

mape_arima = mean_absolute_percentage_error(test, arima_pred["prediction"])

print("ARIMA MAE:", mae_arima)
print("ARIMA RMSE:", rmse_arima)
print("ARIMA MAPE:", mape_arima)
```

```python
mae_sarima = mean_absolute_error(test, sarima_pred["prediction"])

rmse_sarima = np.sqrt(mean_squared_error(test, sarima_pred["prediction"]))

mape_sarima = mean_absolute_percentage_error(test, sarima_pred["prediction"])

print("SARIMA MAE:", mae_sarima)
print("SARIMA RMSE:", rmse_sarima)
print("SARIMA MAPE:", mape_sarima)
```

```python
prophet_values = prophet_pred["yhat"].tail(len(test))
```

```python
mae_prophet = mean_absolute_error(test, prophet_values)

rmse_prophet = np.sqrt(mean_squared_error(test, prophet_values))

mape_prophet = mean_absolute_percentage_error(test, prophet_values)

print("Prophet MAE:", mae_prophet)
print("Prophet RMSE:", rmse_prophet)
print("Prophet MAPE:", mape_prophet)
```

```python
test_lstm = test[-len(lstm_pred):]

mae_lstm = mean_absolute_error(test_lstm, lstm_pred["prediction"])

rmse_lstm = np.sqrt(mean_squared_error(test_lstm, lstm_pred["prediction"]))

mape_lstm = mean_absolute_percentage_error(test_lstm, lstm_pred["prediction"])

print("LSTM MAE:", mae_lstm)
print("LSTM RMSE:", rmse_lstm)
print("LSTM MAPE:", mape_lstm)
```

```python
results = pd.DataFrame({

"Model": ["ARIMA","SARIMA","Prophet","LSTM"],

"MAE":[mae_arima, mae_sarima, mae_prophet, mae_lstm],

"RMSE":[rmse_arima, rmse_sarima, rmse_prophet, rmse_lstm],

"MAPE":[mape_arima, mape_sarima, mape_prophet, mape_lstm]

})

print(results)
```
