```python
import pandas as pd
from prophet import Prophet
```

```python
df = pd.read_csv("data/processed/bitcoin_processed.csv")

df = df[["timestamp","close"]]

df.columns = ["ds","y"]
```

```python
model = Prophet()

model.fit(df)
```

```python
future = model.make_future_dataframe(periods=30)
```

```python
forecast = model.predict(future)
```

```python
prophet_pred = forecast[["yhat"]]

prophet_pred.to_csv("../results/prophet_predictions.csv", index=False)
```

```python
model.plot(forecast)
```
