```python
from textblob import TextBlob
import pandas as pd
import os
```

```python
news_data = [
"Bitcoin price rises as institutional investors enter the market",
"Crypto market crashes after government regulation announcement",
"Ethereum and Bitcoin expected to grow in coming months",
"Investors worried about volatility in cryptocurrency markets",
"Major companies adopting blockchain technology"
]
```

```python
sentiments = []

for news in news_data:
    analysis = TextBlob(news)
    sentiment_score = analysis.sentiment.polarity
    sentiments.append(sentiment_score)
```

```python
df_sentiment = pd.DataFrame({
    "news": news_data,
    "sentiment_score": sentiments
})
```

```python
os.makedirs("../data/raw", exist_ok=True)

df_sentiment.to_csv("data/raw/crypto_sentiment.csv", index=False)

print("Sentiment data saved successfully")
```
