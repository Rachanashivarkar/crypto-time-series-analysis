```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("data/processed/bitcoin_processed.csv")

df["timestamp"] = pd.to_datetime(df["timestamp"])

print(df.head())
```

```python
df.info()

df.describe()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["close"])
plt.title("Bitcoin Price Trend")
plt.xlabel("Date")
plt.ylabel("Price")
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["volume"])
plt.title("Bitcoin Trading Volume")
plt.xlabel("Date")
plt.ylabel("Volume")
plt.show()
```

```python
plt.figure(figsize=(12,6))

plt.plot(df["timestamp"], df["close"], label="Close Price")
plt.plot(df["timestamp"], df["ma7"], label="7-Day MA")
plt.plot(df["timestamp"], df["ma30"], label="30-Day MA")

plt.title("Moving Average Analysis")
plt.legend()
plt.show()
```

```python
plt.figure(figsize=(12,6))
plt.plot(df["timestamp"], df["volatility"])
plt.title("Bitcoin Volatility")
plt.xlabel("Date")
plt.ylabel("Volatility")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(df["returns"], bins=50)
plt.title("Distribution of Daily Returns")
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(), annot=True, cmap="coolwarm")
plt.title("Feature Correlation")
plt.show()
```
